```python
import turtle
wn = turtle.Screen()
wn.bgcolor("lightgreen")
tess = turtle.Turtle()
tess.shape("turtle")
tess.color("blue")

tess.penup()
size = 20
for i in range(30):
    tess.stamp()
    size = size + 3
    tess.forward(size)
    tess.right(24)

tess.color("red")
tess.pendown()
tess.forward(90)

wn.mainloop()
```

```python
import turtle
wn = turtle.Screen()
wn.bgcolor("lightgreen")

clock = turtle.Turtle()
clock.shape("turtle")
clock.color("blue")
clock.speed(1)
clock.pensize(8)

clock.penup()
clock.stamp()

for i in range(12):
    clock.forward(80)
    clock.pendown()
    clock.forward(10)
    clock.penup()
    clock.forward(10)
    clock.stamp()
    clock.back(100)
    clock.right(30)

wn.mainloop()
```
