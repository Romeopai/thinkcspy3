```python
import sys

def sum_to(n):
    """ Return the sum of 1+2+3 ... n """
    ss = 0
    v = 1
    while v <= n:
        ss = ss + v
        v = v + 1
    return ss

def test(did_pass):
    """ Print the result of a test. """
    linenum = sys._getframe(1).f_lineno # Get the caller's line number
    if did_pass:
        msg = "Test at line {0} ok.".format(linenum)
    else:
        msg = ("Test at line {0} FAILED.".format(linenum))
    print(msg)

def test_suite():
    """ Run the suite of tests for code in this module (this file)."""
    test(sum_to(4) == 10)
    test(sum_to(1000) == 500500)

test_suite()
```

```python
def seq3np1(n):
    """ Print the 3n+1 sequence from n, terminating when it reaches 1 """
    while n != 1:
        print(n, end=", ")
        if n % 2 == 0:
            n = n // 2
        else:
            n = n * 3 + 1
    print(n, end=".\n")

print(seq3np1(100))
```

```python
def num_digits(n):
    """ 计算正整数的位数 """
    count = 0
    while n != 0:
        count = count + 1
        n = n // 10
    return count

print(num_digits(4523524))
```

```python
def num_zero_and_five_digits(n):
    count = 0
    while n > 0:
        digit = n % 10 
        if digit == 0 or digit == 5:
            count = count + 1
        n = n // 10
    return count 

print(num_zero_and_five_digits(13455000489))
print(num_zero_and_five_digits(0))
```

```python
def sqrt(n):
    count = 0
    approx = n/2.0
    while True:
        count = count + 1
        better = (approx + n/approx) / 2.0
        if abs(approx - better) < 0.0000001:
            print(count)
            return better
        approx = better

print(sqrt(25.0))
```

```python
def num_digits(n):
    """ 计算任一整数的位数 """
    count = 0
    n = abs(n)

    if n == 0:
        return 1
    else:
        while n != 0:
            count = count + 1
            n = n // 10
        return count

print(num_digits(-24))
print(num_digits(0))
```

```python
def num_even_digits(n):
    count = 0
    if n == 0:
        return 1
    else:
        while n != 0:
            k = n % 10                  # 取余数
            if k % 2 == 0:              # 判断奇数偶数
                count = count + 1       # 计数+1
            n = n // 10                 # 整除，向下取整
        return count

print( num_even_digits(123456) )
print(num_even_digits(2468))
print(num_even_digits(1357))
print(num_even_digits(0))
```

```python
def sum_of_squares(xs):
    sum = 0
    for i in xs:
        sum = sum + i * i
    return sum

print(sum_of_squares([2, 3, 4]))
print(sum_of_squares([ ]))
print(sum_of_squares([2, -3, 4]))

        
```
