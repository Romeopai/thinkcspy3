```python
import turtle
wn = turtle.Screen()
tess = turtle.Turtle()

clrs = ["yellow", "red", "purple", "blue"]
for c in clrs: 
    tess.color(c)
    tess.forward(50)
    tess.left(90)

wn.mainloop()
```

```python
import turtle
wn = turtle.Screen()
wn.bgcolor("lightgreen")
tess = turtle.Turtle()
tess.shape("turtle")
tess.color("blue")

tess.penup()
size = 20
for i in range(30):
    tess.stamp()
    size = size + 3
    tess.forward(size)
    tess.right(24)

tess.color("red")
tess.pendown()
tess.forward(90)

wn.mainloop()
```
