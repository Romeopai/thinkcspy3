```python
import turtle

def draw_bar(t, height):
    t.begin_fill()
    t.left(90)
    t.forward(height)
    t.write(" "+ str(height))
    t.right(90)
    t.forward(40)
    t.right(90)
    t.forward(height)
    t.left(90)
    t.end_fill()
    t.penup()
    t.forward(10)
    t.pendown()

wn = turtle.Screen()
wn.bgcolor("lightgreen")

tess = turtle.Turtle()
tess.color("blue", "red")
tess.pensize(3)

xs = [48, 117, 200, 240, 160, 260, 220]

for v in xs:
    draw_bar(tess, v)

wn.mainloop()
```

```python
def grade_mark(grade):
    if grade >= 75:
        print("First")
    elif (grade >= 70) and (grade < 75):
        print("Upper Second")
    elif (grade >= 60) and (grade <70):
        print("Second")
    elif (grade >= 50) and (grade < 60):
        print("Third")
    elif (grade >= 45) and (grade < 50):
        print("F1 Supp")
    elif (grade >= 40) and (grade < 45):
        print("F2")
    else:
        print("F3")

xs = [83, 75, 74.9, 70, 69.9, 65, 60, 59.9, 55, 50, 49.9, 45, 44.9, 40, 39.9, 2, 0] 

for i in xs:
    grade_mark(i)
```

```python
import turtle

def draw_bar(t, height):
    if (height >= 200):
        t.color("blue", "red")
    elif (height >= 100) and (height < 200):
        t.color("blue", "yellow")
    else:
        t.color("blue", "green")
    t.begin_fill()
    t.left(90)
    t.forward(height)
    t.write(" "+ str(height))
    t.right(90)
    t.forward(40)
    t.right(90)
    t.forward(height)
    t.left(90)
    t.end_fill()
    t.penup()
    t.forward(10)
    t.pendown()

wn = turtle.Screen()
wn.bgcolor("lightgreen")

tess = turtle.Turtle()
tess.pensize(3)

xs = [48, 117, 200, 240, 160, 260, 220, -100, -200]

for v in xs:
    draw_bar(tess, v)

wn.mainloop()
```

```python
def is_rightangled(a, b, c):
    if b < a:
        t = b
        b = a
        a = t
        print(a,b,c)
    
    if c < a:
        t = c
        c = a
        a = t
        print(a,b,c)

    if c < b:
        t = c
        c = b
        b = t
        print(a, b, c)

    if abs( (a**2 + b**2) - c**2 ) < 0.0000001:
        print("This triangle is right angle.")
    else:
        print("This triangle is not right angle.")

is_rightangled(4,5,3)
```
