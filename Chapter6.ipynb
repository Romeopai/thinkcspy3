```python
def bad_absolute_value(x):
    if x < 0:
        return -x
    elif x > 0:
        return x

print(bad_absolute_value(0))
```

> "return" statement means: evaluate the return expression, and then return it immediately as the result (the fruit) of this function.
> 执行return statement后，函数即刻停止循环，返回函数值。

```python
def find_first_2_letter_word(xs):
    for wd in xs:
        if len(wd) == 2:
            return wd
    return "none"

print(find_first_2_letter_word(["This", "is", "a", "dead", "parrot"]))
print(find_first_2_letter_word(["I", "like", "cheese"]))
```

```python
import sys

def day_name(dn):
    if dn == 0:
        return "Sunday"
    elif dn == 1:
        return "Monday"
    elif dn == 2:
        return "Tuesday"
    elif dn == 3:
        return "Wedneday"
    elif dn == 4:
        return "Thursday"
    elif dn == 5:
        return "Friday"
    elif dn == 6:
        return "Saturday"
    else:
        return None

def day_num(dnu):
    if dnu ==  "Sunday":
        return 0
    elif dnu == "Monday":
        return 1
    elif dnu ==  "Tuesday":
        return 2
    elif dnu == "Wednesday":
        return 3
    elif dnu ==  "Thursday":
        return 4
    elif dnu == "Friday":
        return 5
    elif dnu == "Saturday":
        return 6
    else:
        return None

def day_add(dnow, dpass):
    dnu1 = ( day_num(dnow) + dpass ) % 7
    return day_name(dnu1)

def test(did_pass):
    """ Print the result of a test. """
    linenum = sys._getframe(1).f_lineno # Get the caller's line number
    if did_pass:
        msg = "Test at line {0} ok.".format(linenum)
    else:
        msg = ("Test at line {0} FAILED.".format(linenum))
    print(msg)


def test_suite():
    """ Run the suite of tests for code in this module (this file)."""
    test(day_add("Monday", 4) == "Friday")
    test(day_add("Tuesday", 0) == "Tuesday")
    test(day_add("Tuesday", 14) == "Tuesday")
    test(day_add("Sunday", 100) == "Tuesday")
    test(day_add("Sunday", -1) == "Saturday")
    test(day_add("Sunday", -7) == "Sunday")
    test(day_add("Tuesday", -100) == "Sunday")

test_suite()
```

```python
import sys

def to_secs(h, m, s):
    secs = h * 3600 + m * 60 + s
    return int(secs)

def test(did_pass):
    """ Print the result of a test. """
    linenum = sys._getframe(1).f_lineno # Get the caller's line number
    if did_pass:
        msg = "Test at line {0} ok.".format(linenum)
    else:
        msg = ("Test at line {0} FAILED.".format(linenum))
    print(msg)


def test_suite():
    """ Run the suite of tests for code in this module (this file)."""
    test(to_secs(2, 30, 10) == 9010)
    test(to_secs(2, 0, 0) == 7200)
    test(to_secs(0, 2, 0) == 120)
    test(to_secs(0, 0, 42) == 42)
    test(to_secs(0, -10, 10) == -590)
    test(to_secs(2.5, 0, 10.71) == 9010)
    test(to_secs(2.433, 0, 0) == 8758)
test_suite()
```

```python
import sys

def hours_in(secs):
    hours = secs // 3600 # 取整数
    return hours

def minutes_in(secs):
    minutes = ( secs - hours_in(secs) * 3600  ) // 60  # 取整数
    return minutes

def seconds_in(secs):
    seconds = secs - hours_in(secs) * 3600 - minutes_in(secs) * 60
    return seconds

def test(did_pass):
    """ Print the result of a test. """
    linenum = sys._getframe(1).f_lineno # Get the caller's line number
    if did_pass:
        msg = "Test at line {0} ok.".format(linenum)
    else:
        msg = ("Test at line {0} FAILED.".format(linenum))
    print(msg)


def test_suite():
    """ Run the suite of tests for code in this module (this file)."""
    test(hours_in(9010) == 2)
    test(minutes_in(9010) == 30)
    test(seconds_in(9010) == 10)

test_suite()
```

```python
import sys

def slope(x1, y1, x2, y2):
    """求斜率"""
    dy = y2 - y1
    dx = x2 - x1
    return dy/dx

def intercept(x1, y1, x2, y2):
    """求y轴截距"""
    return y2 - slope(x1, y1, x2, y2) * x2

def test(did_pass):
    """ Print the result of a test. """
    linenum = sys._getframe(1).f_lineno # Get the caller's line number
    if did_pass:
        msg = "Test at line {0} ok.".format(linenum)
    else:
        msg = ("Test at line {0} FAILED.".format(linenum))
    print(msg)


def test_suite():
    """ Run the suite of tests for code in this module (this file)."""
    test(slope(5, 3, 4, 2) == 1.0)
    test(slope(1, 2, 3, 2) == 0.0)
    test(slope(1, 2, 3, 3) == 0.5)
    test(slope(2, 4, 1, 2) == 2.0)
    test(intercept(1, 6, 3, 12) == 3.0)
    test(intercept(6, 1, 1, 6) == 7.0)
    test(intercept(4, 6, 12, 8) == 5.0)

test_suite()
```

```python
import sys

def f2c(t):
    """华氏度转摄氏度"""
    c = 5 * (t - 32) / 9
    return round(c, 0) # 返回浮点数的四舍五入值。round(x [, n])

def c2f(t):
    """摄氏度转华氏度"""
    f = 9 * t / 5 + 32
    return round(f, 0)

def test(did_pass):
    """ Print the result of a test. """
    linenum = sys._getframe(1).f_lineno # Get the caller's line number
    if did_pass:
        msg = "Test at line {0} ok.".format(linenum)
    else:
        msg = ("Test at line {0} FAILED.".format(linenum))
    print(msg)

def test_suite():
    """ Run the suite of tests for code in this module (this file)."""
    test(f2c(212) == 100)
    test(f2c(32) == 0)
    test(f2c(-40) == -40)
    test(f2c(36) == 2)
    test(f2c(37) == 3)
    test(f2c(38) == 3)
    test(f2c(39) == 4)
    test(c2f(0) == 32)
    test(c2f(100) == 212)
    test(c2f(-40) == -40)
    test(c2f(12) ==54)
    test(c2f(18) == 64)
    test(c2f(-48) == -54)

test_suite()
```
